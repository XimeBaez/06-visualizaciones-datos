```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

FILENAME = 'tmp/Ejemplo_02_03_Cybersecurity_Dataset.csv'

# Cargar nuestro dataset
cybersecurity_data = pd.read_csv(FILENAME)
cybersecurity_data.head()
```

```python
# Analisis de la informacion
attack_type_counts = cybersecurity_data['attack_type'].value_counts()
print(type(attack_type_counts))
print(attack_type_counts)
```

```python
top_tow_attack_types = attack_type_counts.index[:2]
print(type(top_tow_attack_types))
print(top_tow_attack_types)
```

```python
# Filtro
filtered_data = cybersecurity_data[cybersecurity_data['attack_type'].isin(top_tow_attack_types)]
print(type(filtered_data))
print(filtered_data)
```

# Visualizacion de la informacion | Interpretacion

```python
# Configuracion inicial para seaborn
sns.set(style='darkgrid')

# Configurar scatter
plt.figure(figsize=(10, 6))
sns.scatterplot(data=cybersecurity_data, x='severity', y='response_time', color='b', s=100)

# Configurar titulo y etiquetas
plt.title('Severidad VS Tiempo de respuesta en ataques de ciber seguridad')
plt.xlabel('Severidad del incidente')
plt.ylabel('Tiempo de respuesta')

# mostrar grafico
plt.show()
```

## Visualizacion dos

```python
# Configuracion inicial para seaborn
sns.set(style='darkgrid')

# Configurar scatter
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=filtered_data, 
    x='severity', 
    y='response_time', 
    s=70,
    palette='muted',
    hue='attack_type'
)

# Configurar titulo y etiquetas
plt.title('Severidad VS Tiempo de respuesta en ataques de ciber seguridad')
plt.xlabel('Severidad del incidente')
plt.ylabel('Tiempo de respuesta')

# mostrar grafico
plt.show()
```
